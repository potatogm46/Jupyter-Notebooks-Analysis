```python
Subwaydf = df[df['restaurant'] == 'Subway']
```

```python
tbdf = df[df['restaurant'] == 'Taco Bell']
```

```python
plt.scatter(subwaydf['calories'], subwaydf['cal_fat'])
```

```python
plt.scatter(tbdf['calories'], tbdf['cal_fat'])
```

```python
slope, intercept = np.polyfit(tbdf['calories'], tbdf['cal_fat'], 1)
```

```python
tb_regression_line = slope * tbdf['calories'] + intercept
```

```python
intercept
```

```python
plt.xlim(0, 1300)
plt.ylim(0, 500)
plt.plot(tbdf['calories'], tb_regression_line, color='red')
```

```python
sub_slope, sub_intercept = np.polyfit(subwaydf['calories'], subwaydf['cal_fat'], 1)
sub_regression_line = sub_slope * subwaydf['calories'] + sub_intercept
plt.xlim(0, 1300)
plt.ylim(0, 500)
plt.plot(subwaydf['calories'], sub_regression_line, color='red')
```
